# LOADING DATA


```python
import pandas as pd
import matplotlib.pyplot as plt
crime = pd.read_csv("Crime_Data_from_2020_to_Present.csv")
```

```python
#dropping unwanted columns
df= crime.drop(labels=['modus_operandi','weapon_code','weapon_description','crime_code_2','crime_code_3','crime_code_4','cross_street'] , axis=1)

df
```

```python
# sorting values using a column
df.sort_values(by ="date_reported")
```

```python
df.dtypes
```

```python
# updating data types
df["date_reported"] = pd.to_datetime(df["date_reported"], dayfirst=True, errors="coerce")
df["date_occurred"] = pd.to_datetime(df["date_occurred"], dayfirst=True, errors="coerce")
```

```python
df.info()
```

```python
df.describe(include = ['float64','int64'])
```

```python
df['victim_age'].value_counts()
```

```python
df_age=df[df['victim_age']>0]
df_age['victim_age'].value_counts()
```

```python
#adding 2 more columns
df['year']= df['date_occurred'].dt.year
df['month']= df['date_occurred'].dt.month
```

```python
df
```

#UNDERSTANDING FEATURES


```python
df['month'].value_counts().sort_index()
```

```python
df['month'].value_counts().sort_index().plot()
```

```python
df['year'].value_counts().sort_index().plot(kind ='bar')
```

```python
df=df[~((df['month'] == 12) & (df['year'] == 2023))]
```

```python
df.sort_values(by='date_occurred')
```

```python
df.groupby(['year','month']).size()
```

```python
df.groupby(['year','month']).size().plot()
```

```python
plt.hist(df_age['victim_age'] , edgecolor='black', bins=15)
plt.show()
```

```python
df['date_occurred_no_time'] = df['date_occurred'].dt.date
#dt.date is used to remove time from date column
```

```python
df
```

```python
crime_count_per_day= df.groupby(['date_occurred_no_time','crime_description']).size()
```

```python
crime_count_per_day
```

```python
top_20_crimes = crime_count_per_day.groupby('crime_description').mean()\
                                       .reset_index(name= 'average_daily_crimes').sort_values(by = 'average_daily_crimes' ,ascending = False).head(20)
```

```python
top_20_crimes
```

```python
plt.barh(top_20_crimes['crime_description'] , top_20_crimes['average_daily_crimes'])
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize = (12 , 8))

#plot 1 = line chart
plt.subplot(2,2,1)
crime_plot = df.groupby(['year','month']).size().plot(color="blue" , linewidth=2)
crime_plot.set_xlabel("Year - Month" , fontsize=10)
crime_plot.set_title("Crime over Time")

#plot 2 = histogram
plt.subplot(2,2,2)
plt.hist(df_age['victim_age'] , edgecolor='black', bins=15 , color="lightgreen")
plt.tight_layout()
plt.xlabel("Age" , fontsize=10)
plt.title("Distribution of Victim Age")

#plot 3 = horizontal bar chart
plt.figure()
plt.barh(top_20_crimes['crime_description'] , top_20_crimes['average_daily_crimes'] , color="orange")
plt.title("Top 20 Crimes")
```
